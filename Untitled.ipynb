```python
data = pd.read_excel('Data\Chapter 2\internet-tables-all_test.xls',
                     sheet_name='2019',
                     header=2,
                     usecols='A:AJ')

data.drop(columns='Unnamed: 0', inplace=True) # Drop empty column

data.rename(columns={'2019 (EUR million)':'',
                     'Unnamed: 2': ''}, inplace=True) # Remove wrong labels

data.drop([0, 84, 85, 106, 108, 111, 112, 113, 114], inplace=True) # drop first row

data.reset_index(drop=True, inplace=True) # Reset index
```

```python
data = data.loc[104:105, 'BE':'UK'] # Get relevant rows
#data.columns.tolist()
#data.loc[[104]].values.flatten().tolist()
```

```python
plt.bar(data.columns, (data.loc[[104]]/1000).values.flatten().tolist())
plt.axhline(color='#555555', linewidth=0.8)
plt.title('EU operating budgetary balance, 2019')
plt.ylabel('â‚¬ (billions)')
plt.annotate('Source: European Commission - EU expenditure and revenue 2014-2020', xy=(0, 0), xytext=(0, -.125), xycoords='axes fraction')
plt.show()
```

```python
plt.bar(data.columns, (data.loc[[105]]*100).values.flatten().tolist())
plt.axhline(color='#555555', linewidth=0.8)
plt.title('EU operating budgetary balance, 2019')
plt.ylabel('Percent of gross national income')
plt.annotate('Source: European Commission - EU expenditure and revenue 2014-2020', xy=(0, 0), xytext=(0, -.125), xycoords='axes fraction')
plt.show()
```
